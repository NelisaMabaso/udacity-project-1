# Question 2: Are You Paid More Based on Which Country You Work In?

```python
df_dollar=df[df['Currency']=='U.S. dollars ($)']
df_dollar.dropna(subset=['Salary', 'Country'], inplace=True)
```

```python
# print(df_dollar.groupby('Country')['Salary'].describe())

# Visualizations
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='Salary', data=df_dollar)
plt.xticks(rotation=90)
plt.title('Salary Distribution by Country')
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example of fitting an OLS model
model = ols('Salary ~ C(Country)', data=df_dollar).fit()
print(model.summary())
```

```python
df_dollar['predicted_salary'] = model.predict(df_dollar)
```

```python
# Calculate the average salary by country
salary_by_country = df_dollar.groupby('Country')['Salary'].mean().reset_index()

# Sort the DataFrame by 'Salary' in descending order
salary_by_country_sorted = salary_by_country.sort_values(by='Salary', ascending=False)
```

```python
# Create a pivot table
pivot_table = salary_by_country.pivot_table(index='Country', values='Salary').sort_values(by='Salary', ascending=False)

# Plot heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Average Salary by Country')
plt.savefig('output_image.png')
plt.show()
```

# Question 3: Can we find a ruleset that predict what salary you will have based on all other available features?

```python
# Define salary bins and labels
bins = [0, 30000, 60000, 90000, 120000, 150000, 200000]
labels = ['<30k', '30k-60k', '60k-90k', '90k-120k', '120k-150k', '150k+']

# Create a new column for salary bucket
df_dollar['Salary_Bucket'] = pd.cut(df_dollar['Salary'], bins=bins, labels=labels, right=False)
```

```python
df_dollar.head()
```

```python
# dealing with missing values
# df_dollar = df_dollar.dropna(axis=1, how='all')
# df_dollar=df_dollar.dropna()

# Features and target variable
X = df_dollar.drop(['Salary', 'Salary_Bucket'], axis=1)
# One hot encoding
X = pd.get_dummies(X, drop_first=True)
y = df_dollar['Salary_Bucket']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Initialize and train the decision tree classifier
clf = DecisionTreeClassifier(max_leaf_nodes=10) 
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Classification Report:\n", metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=X.columns, class_names=y.unique(), filled=True, rounded=True)
plt.title('Decision Tree for Predicting Salary Ranges')
plt.show()
```
